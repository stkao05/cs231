```python
import torch
import matplotlib.pyplot as plt
```

```python
feature_size = 200
size = (100, feature_size)
X = torch.randn(size)
```

```python
plt.hist(X.ravel(), 30, range=[-3, 3]);
```

## Init weight too small

We can see that the variance of hidden unit output become smaller as the network goes deeper. This is due to repeated multplication of small weight W

```python
layer_n = 10
H = [0] * layer_n
H[0] = X

for i in range(1, layer_n):
    W = torch.randn((feature_size, feature_size)) * 0.01
    H[i] = (H[i-1] @ W).tanh()
    
plt.figure(figsize=(50, 10))

for i in range(layer_n):
    print(f"H[{i}]: mean: {H[i].mean():.4f}, std:{H[i].std():.4f}")
    plt.subplot(1, layer_n, i+1)
    plt.hist(H[i].ravel(), 30, range=[-3, 3]);
```

When the weight is too large, the tanh would shrink value from the extrame end to be -1 and 1

```python
layer_n = 10
H = [0] * layer_n
H[0] = X

for i in range(1, layer_n):
    W = torch.randn((feature_size, feature_size))
    H[i] = (H[i-1] @ W).tanh()
    
plt.figure(figsize=(50, 10))

for i in range(layer_n):
    print(f"H[{i}]: mean: {H[i].mean():.4f}, std:{H[i].std():.4f}")
    plt.subplot(1, layer_n, i+1)
    plt.hist(H[i].ravel(), 30, range=[-3, 3]);
```

```python
layer_n = 10
H = [0] * layer_n
H[0] = X

for i in range(1, layer_n):
    W = torch.randn((feature_size, feature_size)) / torch.tensor(feature_size).sqrt()
    H[i] = (H[i-1] @ W).tanh()
    
plt.figure(figsize=(50, 10))

for i in range(layer_n):
    print(f"H[{i}]: mean: {H[i].mean():.4f}, std:{H[i].std():.4f}")
    plt.subplot(1, layer_n, i+1)
    plt.hist(H[i].ravel(), 30, range=[-3, 3]);
```
